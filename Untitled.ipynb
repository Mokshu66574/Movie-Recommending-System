```python
final = final.copy()
```

```python
def extract_genre_names(s):
    genre_list = json.loads(s)  # Convert JSON string to list of dicts
    return [genre['name'] for genre in genre_list]  # Extract and lowercase names

final['genre_names'] = final['genres'].apply(extract_genre_names)
```

```python
final.head()
```

```python
final.iloc[0].keywords
```

```python
def name(s1):
    llist = json.loads(s1)  # Convert JSON string to list of dicts
    return [keywords['name'] for keywords in llist]  # Extract and lowercase names

final['new_keywords'] = final['keywords'].apply(name)
```

```python
final.head()
```

```python
def get_cast_names(cast_json):
    cast_list = json.loads(cast_json)
    names = [member["name"] for member in cast_list[:3]]  # Slice first 5 items
    return names
final['new_cast'] = final['cast'].apply(get_cast_names)
```

```python
final.head()
```

```python
def etch(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
final['crew'] = final['crew'].apply(etch)
```

```python
final.head()
```

```python
final = final.dropna(subset=['overview'])
```

```python
final['overview'] = final['overview'].apply(lambda x: x.split())
```

```python
final.head()
```

```python
final['overview'] = final['overview'].apply(
    lambda lst: [word.replace(" ", "") for word in lst]
)
```

```python
final.head()
```

```python
final['new_keywords'] = final['new_keywords'].apply(
    lambda lst: [word.replace(" ", "") for word in lst]
)
```

```python
final['crew'] = final['crew'].apply(
    lambda lst: [word.replace(" ", "") for word in lst]
)
```

```python
final['genre_names'] = final['genre_names'].apply(
    lambda lst: [word.replace(" ", "") for word in lst]
)
```

```python

final['new_cast'] = final['new_cast'].apply(
    lambda lst: [word.replace(" ", "") for word in lst]
)
```

```python
final.head()
```

```python
final['Tags'] = final['overview'] + final['crew'] + final['genre_names'] + final['new_keywords'] + final['new_cast']
```

```python
final.head()
```

```python
full = final[['movie_id' , 'title' , 'Tags']]
```

```python
full.head()
```

```python
full['Tags'] = full['Tags'].apply(lambda x:" ".join(x))
```

```python
full.head()
```

```python
full['Tags'] = full['Tags'].apply(lambda x:x.lower())
```

```python
full.head()
```

```python
data = full.copy()
```

```python
data
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
data['Tags'] = data['Tags'].apply(stem)
```

```python
 cv = CountVectorizer(max_features = 5000 ,  stop_words = "english")
```

```python
vectors = cv.fit_transform(data['Tags']).toarray()
```

```python
vectors
```

```python
# cv.get_feature_names_out()
# for i in range(len(cv.get_feature_names_out())):
#     print(cv.get_feature_names_out()[i])
```

```python
sim = cosine_similarity(vectors)
```

```python
def rocomend(movie):
    movieIndex = data[data['title'] == movie].index[0] 
    distances = sorted(list(enumerate(sim[movieIndex])),reverse = True , key = lambda x:x[1])[1:6]
    for i in distances:
        print(data.iloc[i[0]].title)
```

```python
rocomend('Batman Begins')
```

```python
indices = pd.Series(data.index, index=data['title']).to_dict()
```

```python
import joblib

# Assuming your recommendation function is called "recomend" and uses these components:
model_data = {
    'movie_data': data,  # Your movie dataset
    'similarity_matrix': sim,  # If you're using a similarity matrix
    'indices': indices  # If you have movie title to index mapping
}

# Export the model data
joblib.dump(model_data, 'movie_recommendation_model.pkl')
```
